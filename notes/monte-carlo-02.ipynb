```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 標準正規分布から10,000個の値を生成し、20を掛ける
values = np.random.randn(10000) * 20

# パーセンタイル値 (x) を定義
x = 1  # xパーセンタイル

# xパーセンタイル値を計算
percentile_value = np.percentile(values, x)
mean_value = np.mean(values)
std_deviation = np.std(values)
print(f"平均値は: {mean_value}, 標準偏差は: {std_deviation}")
print(f"{x}パーセンタイルの値は: {percentile_value}")
```

```python
# 度数分布のグラフを描画
plt.figure(figsize=(10, 6))
plt.hist(values, bins=50, edgecolor='black', alpha=0.7)
plt.title('Histogram of Generated Random Values')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# 定数kの値を設定
k = 2.5

# 期待値と標準偏差の組み合わせを2種類用意
params = [(mean_value, std_deviation), (0, 1)]

# グラフを描画
plt.figure(figsize=(12, 6))

# グラフを2つに分けて並べて表示する
fig, axs = plt.subplots(1, 2, figsize=(18, 6))

for i, (mu, sigma) in enumerate(params):
    x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
    pdf = (1/(sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma)**2)
    axs[i].plot(x, pdf, label=f'μ={mu}, σ={sigma}')
    axs[i].fill_between(x, 0, pdf, where=(x <= mu + k * sigma), alpha=0.3)
    axs[i].set_title(f'Normal Distribution with μ={mu}, σ={sigma}')
    axs[i].set_xlabel('Value')
    axs[i].set_ylabel('Probability Density')
    axs[i].legend()
    axs[i].grid(True)

plt.show()
```
