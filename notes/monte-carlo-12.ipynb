## リスク中立経済下における期待値

(「どうやったらリスク中立経済下における期待値が求められるか」の説明)  

### 期待値（リスクフリーレートから）

ポイントは前々回の「リスク中立経済においては金融商品のリターンの期待値はリスクフリーレートと同じ」。  
つまり、
- オプション満期までの期間に対応するリスクフリーレートを r （ただし連続複利ベース）
- 時点 T における金融商品の価格を ST
- そのリスク中立経済下における期待値 $E^{*}(S_{T})$ は以下の式が成り立つ
  - （Ｅの右肩についている ＊ は「リスク中立経済下における期待値」を示す特殊表現） 

$E^{*}(S_{T}) = S_{0}e^{rT}$

### 期待値（原資産の価格変化モデルから）

また、$S_T = S_{0}e^{\mu T+{\sigma}W_{T}}$ ・・(2)式から期待値は以下の値となる

$E^*(S_T) = S_{0}e^{{\mu}T+\frac{\sigma^2T}{2}}$
- ※導出
- 対数正規分布に従う
  - $S_T = S_{0}e^{\mu T+{\sigma}W_{T}}$
  - $\ln(S_T) = \ln(S_0) + \mu T + \sigma W_T$
    - $\ln(S_T) = \ln(S_0) + \mu T$ は定数、$\sigma W_T$ は正規分布
    - →　$\ln(S_T)$ は正規分布。
    - →　$S_T$ は対数正規分布。
    - 対数正規分布に従う場合は期待値の公式があるらしい
      - $E[Y] = e^{\mu + \frac{\sigma ^2}{2}}$
- 期待値
  - $E[S_T] = E[S_{0}e^{\mu T+{\sigma}W_{T}}]$
  - $E[S_T] = S_{0}E[e^{\mu T+{\sigma}W_{T}}]$
  - $E\left[e^{uT + \sigma W_T}\right] = e^{uT} \cdot E\left[e^{\sigma W_T}\right]$
    - 正規分布における指数関数の期待値の公式を使って
    - $E[e^{\sigma W_T}] = e^{\mu W + \frac{\sigma^2 W}{2}}$
      - $\mu W = 0$  （正規分布  $W_T$  の平均 = 0）
      - $\sigma^2 W = \sigma^2 T$  （正規分布  $\sigma W_T$  の分散）
    - $E[e^{\sigma W_T}] = e^{0 + \frac{\sigma^2 T}{2}} = e^{\frac{\sigma^2 T}{2}}$
    - $E\left[e^{uT + \sigma W_T}\right] = e^{uT} \cdot e^{\frac{\sigma^2 T}{2}} = e^{uT + \frac{\sigma^2 T}{2}}$
  - $E[S_T] = S_{0}E[e^{\mu T+{\sigma}W_{T}}]$
  - $E[S_T] = S_0 e^{\mu T + \frac{\sigma^2 T}{2}}$

### 期待値から$\mu$を求める

リスクフリーレートおよび原資産の価格変化モデルの２つの期待値は一致するので

- $S_{0}e^{{\mu}T+\frac{\sigma^2T}{2}} = S_{0}e^{rT}$
- ${\mu}T+\frac{\sigma^2T}{2} = {rT}$
- ${\mu}T = {rT} - \frac{\sigma^2T}{2}$
- ${\mu} = {r} - \frac{\sigma^2}{2}$

求められた $\mu$ を $S_T = S_{0}e^{\mu T+{\sigma}W_{T}}$ に代入してシミュレーションをすると「リスク中立経済を前提にしたシミュレーション」になる

## ちょっとややこしい話

`リスク中立経済下においては、金融商品のリターンの期待値がリスクフリーレートになる必要があることから、ある金融商品 S の時点 T における価値の期待値は、以下の式を満たす必要`

$E^{*}(S_{T}) = S_{0}e^{rT}$

と記載していた。

しかし、為替レートを対象にしたオプションの場合にはそのままでは当てはまらない。

具体例）
- 現時点で１ドルを 80 円で購入した
- １年後、為替レートが１ドル 90 円になっていたとすると、１年前に買ったドルの価値は 90 円か？
- **ドルの金利で運用**できるので、ドルの１年物金利を q とすれば ($(1 + q) * 90$ )円 になる


**金融商品の価格を理論的に考えるときは、本来運用して増やせるものは当然運用するという考え方を前提**とする
- リスク中立経済下における１年後の為替レートの期待値
  - （ドル金利で1年間運用した１ドルの価値×1年後の為替レート）の期待値が「現在の為替レート× $e^r$ 」になる

以上を式で表現すると
- 前提
  - 現時点でのドル円為替レートを $S_0$
  - １年後のドル円為替レートを $S_1$
  - １年物円金利（ただし連続複利）を $r$
  - ドル金利を $q$
- **リスク中立経済下における $S_1$ の期待値**
  - $E^{*}(S_{1}) = S_{0}e^{r}$ ではなく、ドル運用の$e^q$を追加して
  - $E^{*}(S_{1}e^q) = S_{0}e^{r}$ → $E^{*}(S_{1}) = S_{0}e^{r-q}$
  - $S_{0}e^{{\mu}T+\frac{\sigma^2T}{2}} = S_{0}e^{rT}$ に当てはめると $S_{0}e^{{\mu}T+\frac{\sigma^2T}{2}} = S_{0}e^{(r-q)T}$
    - ${\mu}T+\frac{\sigma^2T}{2} = {(r-q)T}$
    - ${\mu} = r - q - \frac{\sigma^2T}{2}$
  - ${\mu}$ を利用してシミュレーションを行い`Max（満期時為替レート － 行使価格 , 0）`の期待値を出す
- (最後にこれを現在価値に直す)



## BS式とシミュレーション

### 前提

- コールオプション
- 満期　：　１年
- 現在のドル／円為替レート　：　1ドル ＝ 80 円
- 行使価格　：　80 円
- 円金利　：　0.2％
- ドル金利　：　0.5％
- ボラティリティ　：　10％


### シミュレーション

${\mu} = r - q - \frac{\sigma^2T}{2}$

$S_T = S_{0}e^{uT+{\sigma}W_{T}}$

$S_1 = S_{0}e^{u1+{\sigma}W_{T}}$

```python
import numpy as np

d_rate = 0.005
print("ドル金利 は " + str(d_rate))

y_rate = 0.002
print("円金利 は " + str(y_rate))

m = 0.002 - 0.005 - (0.1 ** 2) / 2
print("mu は " + str(m))

S0 = 80
print("S0 は " + str(S0))

Strike = 80
print("行使価格 は " + str(Strike))

v = 0.1
print("ボラティリティ は " + str(v))

sim_count = 5000000
print("シミュレーション回数 は " + str(sim_count))
# 標準正規分布に従う乱数をsim_count個生成
Wt = np.random.normal(0, 1, sim_count)
# 乱数の先頭5個を表示
print("乱数のサンプル" + str(Wt[:5]))
# Wtにvを乗算する
sWt = Wt * v
# sWtの先頭5個を表示
print("ボラリティ乗算後のサンプル" + str(sWt[:5]))

# sWtの要素ごとに、(S0 + 自然定数eの(m + sWt)乗)を計算
St = S0 * np.exp(m + sWt)
# Stの先頭5個を表示
print("Stのサンプル" + str(St[:5]))
print("Stの平均は " + str(np.mean(St)))

# StがStrikeよりも大きい場合はSt、小さい場合は0を設定する
price = np.maximum(St - Strike, 0)
# priceの先頭5個を表示
print("価値のサンプル" + str(price[:5]))

# priceの平均を計算
sim_result = np.mean(price)
print("価値の平均は " + str(sim_result))

# priceの平均を円金利で割る
# 13回の結果から円金利で計算していそうなので、円金利で割るのが正しいと思われる
sim_result = np.mean(price) / (1 + y_rate)
print("価値の平均を円金利で割ると " + str(sim_result))
```

```python
# Python3ではじめるシステムトレード：ブラック-ショールズ方程式を理解する
# https://qiita.com/innovation1005/items/98e28dd0e38246d1dbac

from scipy.stats import norm
import numpy as np

def ecall(s,k,t,r,v):
    d1=(np.log(s/k)+(r+v**2/2)*t)/v/np.sqrt(t)
    d2=(np.log(s/k)+(r-v**2/2)*t)/v/np.sqrt(t)
    return s*norm.cdf(d1)-k*np.exp(-r*t)*norm.cdf(d2)

bs_result = ecall(80, 80, 1, -0.003, 0.1)
print("BS式では " + str(bs_result))
# BS式の結果を円金利で割る
#bs_result = bs_result / (1 + y_rate)
#print("BS式の結果をドル金利で割ると " + str(bs_result))


```

```python
# シミュレーションとBS式の結果を比較する
print("シミュレーションとBS式の結果を比較する")
print("シミュレーションの結果は " + str(sim_result))
print("BS式の結果は " + str(bs_result))
print("シミュレーションの結果とBS式の結果の差は " + str(sim_result - bs_result))
```
