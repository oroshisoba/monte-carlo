### プライシングの基本的考え方

`『デリバティブ商品の価格は、リスク中立経済下における当該デリバティブ商品の将来価値の期待値』の現在価値`  
(「デリバティブ商品の価格は、『リスク中立経済下における当該デリバティブ商品の将来価値の期待値』の現在価値」じゃないのだろうか？)


#### 満期時のコールオプションの価値

`Max（満期時為替レート － 行使価格 , 0）`  ・・・（１）式  
(満期時に為替レートが行使価格よりも高いときにだけ、コールオプションを行使する価値がある)

#### 原資産の価格変化モデル式

$S_t = S_{0}e^{uT+{\sigma}W_{T}}$  ・・・（２）式  

1. このモデル式にしたがってドル円為替レートが変化すると仮定
2. リスク中立経済下における（１）式の値の期待値を求める
3. 結果を現在価値にする


**どうやったらリスク中立経済下における期待値が求められるか**がオプション・プライシング上の非常に重要なキモの部分。

## リスク中立経済下における期待値

(「どうやったらリスク中立経済下における期待値が求められるか」の説明)  

ポイントは前々回の「リスク中立経済においては金融商品のリターンの期待値はリスクフリーレートと同じ」。
つまり、オプション満期までの期間に対応するリスクフリーレートを r （ただし連続複利ベース）とすると、 
時点 T における金融商品の価格を ST としたとき、そのリスク中立経済下における期待値 $E^{*}(S_{T})$ （Ｅの右肩についている ＊ は「リスク中立経済下における期待値」を示す特殊表現） 
は以下の値でなければなりません。

$E^{*}(S_{T}) = S_{0}e^{rT}$

また、(2)式から期待値は以下になります。

$S_{0}e^{{\mu}T+\frac{\sigma^2T}{2}}$

期待値は一致するので

$S_{0}e^{{\mu}T+\frac{\sigma^2T}{2}} = S_{0}e^{rT}$

${\mu}T+\frac{\sigma^2T}{2} = {rT}$

${\mu}T = {rT} - \frac{\sigma^2T}{2}$

${\mu} = {r} - \frac{\sigma^2}{2}$

求められた$\mu$を(2)式に代入してシミュレーションをすれば、「リスク中立経済を前提にしたシミュレーション」を行っていることになる。

## ちょっとややこしい話

評価対象が為替レートを対象にしたオプションであるために、さらに「ひとひねり」必要

`リスク中立経済下においては、金融商品のリターンの期待値がリスクフリーレートになる必要があることから、ある金融商品 S の時点 T における価値の期待値は、以下の式を満たす必要`

$E^{*}(S_{T}) = S_{0}e^{rT}$

しかしこれは為替レートにはそのままあてはまらない

具体例）
- 現時点で１ドルを 80 円で購入した
- １年後、為替レートが１ドル 90 円になっていたとすると、１年前に買ったドルの価値は 90 円か？
- ドルの金利で運用できるので、ドルの１年物金利を q とすれば ($(1 + q) * 90$ )円 になる


- 金融商品の価格を理論的に考えるときは、本来運用して増やせるものは当然運用するという考え方を前提とする
- リスク中立経済下における１年後の為替レートの期待値
  - （ドル金利で1年間運用した１ドルの価値×1年後の為替レート）の期待値が「現在の為替レート× $e^r$ 」になるという形で決まる（〜〜という条件を満たすように決定されるってこと？）

以上を式で表現すると
- 前提
  - 現時点でのドル円為替レートを $S_0$
  - １年後のドル円為替レートを $S_1$
  - １年物円金利（ただし連続複利）を $r$
  - ドル金利を $q$
- リスク中立経済下における $S_1$ の期待値
  - $E^{*}(S_{1}) = S_{0}e^{r}$ ではなく、ドル運用の$e^q$を追加して
  - $E^{*}(S_{1}e^q) = S_{0}e^{r}$ → $E^{*}(S_{1}) = S_{0}e^{r-q}$
  - $S_{0}e^{{\mu}T+\frac{\sigma^2T}{2}} = S_{0}e^{rT}$ に当てはめると $S_{0}e^{{\mu}T+\frac{\sigma^2T}{2}} = S_{0}e^{(r-q)T}$
    - ${\mu}T+\frac{\sigma^2T}{2} = {(r-q)T}$
    - ${\mu} = r - q - \frac{\sigma^2T}{2}$
  - ${\mu}$ を利用してシミュレーションを行い`Max（満期時為替レート － 行使価格 , 0）`の期待値を出す
