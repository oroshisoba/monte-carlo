```python
import random

# 0から1までの乱数を生成
random_number = random.random()
print(random_number)

# 1から10までの整数の乱数を生成
random_integer = random.randint(1, 10)
print(random_integer)
```

```python
import numpy as np
# 生成数
num_small = 1000
num_large = 20000

# 一様乱数を生成
uniform_random_numbers_1000 = np.random.uniform(0, 1, num_small)
uniform_random_numbers_10000 = np.random.uniform(0, 1, num_large)

# ヒストグラムをプロット
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(uniform_random_numbers_1000, bins=30, density=True, alpha=0.6, color='g')
plt.title('Uniform Random Numbers PDF (n = ' + str(num_small) + ')')
plt.xlabel('Value')
plt.ylabel('Density')

plt.subplot(1, 2, 2)
plt.hist(uniform_random_numbers_10000, bins=30, density=True, alpha=0.6, color='g')
plt.title('Uniform Random Numbers PDF (n =' + str(num_large) + ')')
plt.xlabel('Value')
plt.ylabel('Density')

# グラフを表示
plt.tight_layout()
plt.show()

```

```python
# 生成数
num_small = 200
num_large = 20000

# 標準正規乱数を生成
normal_random_numbers_1000 = np.random.randn(num_small)
normal_random_numbers_10000 = np.random.randn(num_large)

# ヒストグラムをプロット
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(normal_random_numbers_1000, bins=30, density=True, alpha=0.6, color='b')
plt.title('Standard Normal Random Numbers PDF (n = ' + str(num_small) + ')')
plt.xlabel('Value')
plt.ylabel('Density')

plt.subplot(1, 2, 2)
plt.hist(normal_random_numbers_10000, bins=30, density=True, alpha=0.6, color='b')
plt.title('Standard Normal Random Numbers PDF (n = ' + str(num_large) + ')')
plt.xlabel('Value')
plt.ylabel('Density')

# グラフを表示
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import norm

def probability_below(d):
    """
    標準正規分布に従う確率変数が値 d 以下の値をとる確率を計算する関数
    """
    return norm.cdf(d)

# 例: d = 0 の場合の確率を計算
d = 0
prob = probability_below(d)
print(f"標準正規分布に従う確率変数が値 {d} 以下の値をとる確率: {prob}")
```

```python
# 標準偏差を変えた正規乱数を生成
std_dev_1 = 1
std_dev_2 = 2
mean = 0
num_samples = 10000

normal_random_numbers_std1 = np.random.normal(mean, std_dev_1, num_samples)
normal_random_numbers_std2 = np.random.normal(mean, std_dev_2, num_samples)

# ヒストグラムをプロット
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(normal_random_numbers_std1, bins=30, density=True, alpha=0.6, color='r')
plt.title(f'Normal Distribution (mean = {mean}, std = {std_dev_1})')
plt.xlabel('Value')
plt.ylabel('Density')

plt.subplot(1, 2, 2)
plt.hist(normal_random_numbers_std2, bins=30, density=True, alpha=0.6, color='r')
plt.title(f'Normal Distribution (mean = {mean}, std = {std_dev_2})')
plt.xlabel('Value')
plt.ylabel('Density')

# グラフを表示
plt.tight_layout()
plt.show()
```
